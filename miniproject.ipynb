```python
Data=pd.read_csv("/content/50_Startups.csv")
x=Data['Administration']
y=Data['State']
x1=Data['R&D Spend']
x2=Data['Administration']
x3=Data['Marketing Spend']
x4=Data['State']
x5=Data['Profit']
```

```python
#Linear Regression
import pandas as pd
import numpy as np

m_values=[]
c_values=[]
error_rate=[]

def LinearReg(m1,m2,m3,m4,m5,c):
  yp=[]
  ypp=[]
  
  m_x=[]
  for i in range(0,len(x1)):
    yp.append(m1*x1[i]+m2*x2[i]+m3*x3[i]+m4*x4[i]+m5*x5[i]+c)
    ypp.append((y[i]-yp[i])**2)
  sum=np.array(ypp).sum()
  error_rate.append(sum/float(len(x1)))
```

```python
m=0
c=0
import random
m_values.append(1)
c_values.append(1)
for i in range(0,600):
  a=random.randint(0,20)
  LinearReg(i+a,a+a,i,i+1,i+a,a)
  #LinearReg((random.randint(0,12),))

print("error rates=",error_rate)
min_error=np.array(error_rate).min()
min_m=error_rate.index(min_error)
print("minimum error obtained at m =",m_values[min_m])
m_best=m_values[min_m]
c_best=c_values[min_m]
print("minimum error obntained at c =",c_values[min_m])
print("minimum error =",min_error)

iterate=np.arange(0,600)
plt.plot(iterate,error_rate)
```
